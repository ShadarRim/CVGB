```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import imageio
```

```python
img = imageio.imread('cat-bw.png')[:,:,0]
print(img.shape)
```

```python
img
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img, cmap='gray')
```

```python
def inversion(x):
    out_img = np.zeros((x.shape[0], x.shape[1]), dtype=np.uint8)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            out_img[i, j] = 255 - x[i, j]
    return out_img
```

```python
res = inversion(img)
plt.figure(figsize=(8, 8))
plt.imshow(res, cmap='gray')
```
